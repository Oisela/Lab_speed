```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

data = "data/Fr_02_PS4.csv"

# read CSV into a structured NumPy array (assumes first row is header)
data = np.genfromtxt(data, delimiter=',', names=True)

# split into individual columns for easier use
time = data['Time_since_start_of_measurement']
hall_voltage = data['Hallspannung']
long_voltage = data['Längsspannung']
temperature = data['Temperatur']

x = temperature
y = hall_voltage

# plot the data points
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='None', label='p-dotiert')
plt.xlabel(f"Temperatur in °C")
plt.ylabel(f'$U_H$ in mV')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = temperature
y = long_voltage

# plot the data points
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='None', label='p-dotiert')
plt.xlabel(f"Temperatur in °C")
plt.ylabel(f'Längsspannung in V')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = temperature
y = long_voltage / (25*10**-3)

# plot the data points
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='None', label='p-dotiert')
plt.xlabel(f"Temperatur in °C")
plt.ylabel(f'Wiederstand in $\Omega$')
plt.grid(True)
plt.legend()
plt.show()
```
