```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

x = np.array([5,10,15,20,25,30])
y = np.array([-7.8,-15.1,-22.9,-30.8,-38.1,-45.9])

# plot the data points
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='None', label='n-dotiert')


# compute linear fit
slope, intercept = np.polyfit(x, y, 1)
y_fit = slope * x + intercept

# compute R-squared
residuals = y - y_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y - np.mean(y))**2)
r2 = 1 - ss_res/ss_tot

# plot the fit line with R² in label
plt.plot(
    x, y_fit, color='red', linestyle='-',
    label=f'Fit: f(x) = {slope:.2f}x + {intercept:.2f}, $R^2$ = {r2:.4f}'
)
plt.xlabel(f"Probestrom in mA")
plt.ylabel(f'$U_H$ in mV')
plt.grid(True)
plt.legend()


plt.show()
```
